### Dependencies and Setup

```python

import csv
import os
import time
import random
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
from scipy.stats import linregress
from pprint import pprint
import numpy as np
import time
import datetime
import scipy.stats as st
import gmaps

# Import API key
from api_keys import weather_api_key
from api_keys import g_key
```

## Load Part I results into DataFrame
Load the csv exported in Part I to a DataFrame

```python
# read the clean_cities_output.csv generated by the WeatherPy.ipynb to create dataframe, previous cells and api calls to citpy and open weather map api
# are not needed to be executed as long as citclean_cities_output.csv was created
clean_cities = pd.read_csv("../input_data/clean_cities_output.csv", encoding="utf-8")
clean_cities.head()
```

### Humidity Heatmap
1. Configure gmaps.
2. Use the Lat and Lng as locations and Humidity as the weight.
3. Add Heatmap layer to map.

```python
humidity = clean_cities["Humidity"].astype(float)
max_humidity = humidity.max()

print(max_humidity)
clean_city_locations = clean_cities[["Lat", "Lng"]]
```

```python
# Configure gmaps with API key
gmaps.configure(api_key=g_key)
```

```python
# Create gmaps figure objects
fig = gmaps.figure(center=(0.0, 0.0), zoom_level=2)

heat_layer = gmaps.heatmap_layer(clean_city_locations, weights=humidity,
                                 max_intensity=max_humidity,
                                 dissipating=False,
                                 point_radius=3)
fig.add_layer(heat_layer)
fig
```

#### Create selected cities DataFrame with a very pleasent weather criteria
1. Select cities to fit pleasent weather conditions.
2. Drop any rows will null values.

```python
selected_clean_city_df = clean_cities.loc[(clean_cities["Max Temp"] > 70) &
                                          (clean_cities["Max Temp"] < 80) &
                                          (clean_cities["Wind Speed"] < 10) &
                                          (clean_cities["Cloudiness"] == 0)
                                          , :]
selected_clean_city_df = selected_clean_city_df.dropna(how='any')
selected_clean_city_df.reset_index(inplace=True)
del selected_clean_city_df['index']

# Display to output
selected_clean_city_df.head()
```

### Hotel Map
1. Store into variable named hotel_df.
2. Add a "Hotel Name" column to the DataFrame.
3. Set parameters to search for hotels with 5000 meters.
6. Retreive the Google Places API for each city's coordinates.
7. Store the first Hotel result into the DataFrame.
9. Plot markers on top of the heatmap.

```python
hotel_list = []

for i in range(len(selected_clean_city_df)):
    lat = selected_clean_city_df.loc[i]['Lat']
    lng = selected_clean_city_df.loc[i]['Lng']

    params = {
        "location": f"{lat},{lng}",
        "radius": 5000,
        "types" : "hotel",
        "key": g_key
    }
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    requested = requests.get(base_url, params=params)
    jsn = requested.json()
    try:
        hotel_list.append(jsn['results'][0]['name'])
    except:
        hotel_list.append("")

# Append Hotel Name to the selected clean city data frame
selected_clean_city_df["Hotel Name"] = hotel_list
selected_clean_city_df = selected_clean_city_df.dropna(how='any')
selected_clean_city_df.head()
```

```python
# Create hotel subset data set

selected_clean_city_hotel_df = selected_clean_city_df[["Hotel Name", "City", "Country"]]

selected_clean_city_hotel_df.head()
```

```python
# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
hotel_info = [info_box_template.format(**row) for index, row in selected_clean_city_hotel_df.iterrows()]
hotel_locations = selected_clean_city_df[["Lat", "Lng"]]
```

```python
marker_layer = gmaps.marker_layer(hotel_locations, info_box_content=hotel_info)
fig = gmaps.figure(center=(0.0, 0.0), zoom_level=2)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

fig
```
